```python
# 수학의 행렬의 곱연산
import numpy as np

A = np.array([
    [1, 2],
    [3, 4]
])

B = np.array([
    [5, 6],
    [7, 8]
])

# 수학의 행렬의 곱연산 
# 1×5 + 2×7) = 19
# (1×6 + 2×8) = 22
# (3×5 + 4×7) = 43
# (3×6 + 4×8) = 50

print(np.dot(A, B)) # or print(A @ B) 같은 연산
```

```python
# NumPy에서는 수학의 행렬 곱을 그대로 @ 또는 dot으로 계산할 수 있다.
print(A @ B)
```

```python
# 데이터 분석에서 행렬과 백터의 곱이 필요한 이유
import numpy as np

X = np.array([
 [2, 80],
 [3, 70],
 [1, 90],
 [4, 60],
 [5, 85],
 [2, 75],
 [3, 95],
 [6, 50],
 [4, 88],
 [5, 92]
])

w = np.array([0.5, 0.01])

y = X @ w

print(y) # 이건 10명의 사람 각각에 대한 결과값입니다.
```
