```python
import pymongo
from pymongo import MongoClient
```

```python
#1.MongoClient 를 연결
client = MongoClient('localhost', 27017)
# client = MongoClient('mongodb://localhost:27017')
```

```python
#2. 데이터베이스 연결
# 데이터베이스가 없다면, 새롭게 생성
# 데이터베이스가  있다면, 연결
# sample 이름의 데이터베이스가 생성 됨
db = client.sample
```

```python
#3. 컬렉션 연결
# 컬렉션이 없다면, 새롭게 생성
# 컬렉션이 있다면, 연결
# test 이름의 컬렉션이 생성됨
collection = db.test
```

### insert 
- document insert
- insert() : deprecated
- insert_one()
- insert_many()

```python
print(pymongo.__version__)
```

```python
collection.insert_one({'이름':'이진욱', '직업':'IT개발자'})
```

```python
#collection.insert({'이름':'현빈', '직업':'배우'})
```

```python
collection.insert_many( [ {'이름':'표선영', '직업':'강사'},
                       {'이름':'현빈', '직업':'IT운영자'},
                       {'이름':'주지훈', '직업':'배우'},
                       {'이름':'김남길', '직업':'IT운영자'},
                       {'이름':'손예진', '직업':'IT개발자'} ] )  
```

### find(select)
- find()
- find_one()

```python
for item in collection.find():
    print(item)
```

```python
collection.find_one({'이름':'표선영'})
```

```python
collection.find_one({'이름':'현빈'})
```

```python
for item in collection.find().sort('이름'): 
    print(item)
```

```python
for item in collection.find().sort('이름', pymongo.DESCENDING): 
    print(item)
```

```python
for item in collection.find().sort('이름', pymongo.ASCENDING): 
    print(item)
```

### update
- update() : deprecated
- update_one()
- update_many()

```python
collection.update_one({'이름':'이진욱'},{ '$set': {'직업':'IT운영자'} })
```

```python
for item in collection.find().sort('이름'): 
    print(item) 
```

```python
collection.update_many({'직업':'배우'},{ '$set': {'직업':'연예인'} })
```

```python
for item in collection.find().sort('이름'): 
    print(item) 
```

### delete
